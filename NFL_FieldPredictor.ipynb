```python
import pandas as pd
import numpy as np
from sklearn import preprocessing, cross_validation, svm
```

```python
from glob import iglob
path = "CSVs/*.csv"

df = pd.concat((pd.read_csv(f) for f in iglob(path, recursive=True)), ignore_index=True)
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
df.drop(['qb'], 1, inplace=True)
df.drop(['lg'], 1, inplace=True)

df.head(10)
```

```python
df = df[df["att"] >= 10]
df = df[df["rate"] >= 10]
df.tail(10)
```

```python
dictMane = {'away': int(0), 'home': int(1)}

df['home_away'] = df['home_away'].map(dictMane)
```

```python
df.head(10)
```

```python
def normalizeArr(arr):
    normMin = 10.1
    normMax = 158.3
    diff = normMax - normMin
    for index,num in enumerate(arr):
        arr[index] = (num - normMin)/(diff)
    return arr
```

```python
def normalizeInput(num):
    normMin = 10.1
    normMax = 158.3
    diff = normMax - normMin
    retNum = (num - normMin)/(diff)
    return retNum
```

```python
df["rate"] = normalizeArr(np.array(df["rate"]))
```

```python
df.head(5)
```

```python
df = df[df["game_points"] >= 3]
df = df[df["game_points"] <= 60]
```

```python
def normalizeGamePoints(num):
    normMin = 3
    normMax = 60
    diff = normMax - normMin
    retNum = (num - normMin)/(diff)
    return retNum
```

```python
df["game_points"] = normalizeGamePoints(np.array(df["game_points"]))
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
# Only using QB rating because all of the other factors such as attempts, completions, total yards, yards per attempt, 
# touchdowns, interceptions, etc. are taken into account to get the final QB passer rating.
paramOne = np.array(df['rate'])
paramTwo = np.array(df['game_points'])
X = np.stack((paramOne, paramTwo), axis = 1)
y = np.array(df['home_away'])
```

```python
X
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2)
```

```python
# real_x_train = X_train.reshape(-1, 1)
# real_x_test = X_test.reshape(-1, 1)
```

```python
clf = svm.SVC()
clf.fit(X_train, y_train)
```

```python
accuracy = clf.score(X_test, y_test)
print(accuracy)
```

Accuracy is 50.95% so it doesn't look like there is a direct correlation between QB passer rating of the game, and the final points VS if the game was played in the home field or away. Probability is basically like flipping a coin!
