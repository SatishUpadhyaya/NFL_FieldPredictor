```python
import pandas as pd
import numpy as np
from sklearn import preprocessing, cross_validation, svm
```

```python
df = pd.read_csv('CSVs/QBStats_2010.csv')
df2 = pd.read_csv('CSVs/QBStats_2011.csv')
df3 = pd.read_csv('CSVs/QBStats_2012.csv')
df4 = pd.read_csv('CSVs/QBStats_2013.csv')
df5 = pd.read_csv('CSVs/QBStats_2014.csv')
df6 = pd.read_csv('CSVs/QBStats_2015.csv')
df7 = pd.read_csv('CSVs/QBStats_2016.csv')

df = df.append(df2)
df = df.append(df3)
df = df.append(df4)
df = df.append(df5)
df = df.append(df6)
df = df.append(df7)
```

```python
df.head(10)
```

```python
df.drop(['qb'], 1, inplace=True)
df.drop(['lg'], 1, inplace=True)

df.head(10)
```

```python
dictMane = {'away': int(0), 'home': int(1)}

df['home_away'] = df['home_away'].map(dictMane)
```

```python
df.head(10)
```

```python
# Only using QB rating because all of the other factors such as attempts, completions, total yards, yards per attempt, 
# touchdowns, interceptions, etc. are taken into account to get the final QB rating.
X = np.array(df['rate'])
y = np.array(df['home_away'])
```

```python
#X
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2)
```

```python
real_x_train = X_train.reshape(-1, 1)
real_x_test = X_test.reshape(-1, 1)
```

```python
clf = svm.SVC()
clf.fit(real_x_train, y_train)
```

```python
accuracy = clf.score(real_x_test, y_test)
print(accuracy)
```

Accuracy is 52.15% so it doesn't look like there is a direct correlation between QB rating of the game VS if the game was played in the home field or away. Probability is basically like flipping a coin!
